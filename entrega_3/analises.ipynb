```python
import pandas as pd

files = list(range(1,9))

for file in files:
    # Carrega o CSV
    df = pd.read_csv(f'results/test_0{file}/experiments_results_test.csv')

    # Agrupa por configuração
    grouped = df.groupby(['selection', 'penalty', 'n_gen', 'pop_size'])

    # Calcula estatísticas
    summary = grouped['best_value'].agg([
        ('mean', 'mean'),
        ('std', 'std'),
        ('best_value', 'max'),
        ('worst_value', 'min')
    ]).reset_index()

    # Salva os resultados em um novo CSV
    summary.to_csv(f'results/test_0{file}/summary_results.csv', index=False)

    # Exibe as 10 primeiras linhas
    print("Resumo dos resultados (média, desvio padrão, melhor e pior valor):")
    print(summary.head(10))
```
